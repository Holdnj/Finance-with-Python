Plotting with Yahoo Finance

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
from datetime import date
import mplfinance as mpf

sd = date(2020, 1, 1)
ed = date(2025, 1, 1)

AMD = yf.download("AMD", start=sd, end=ed, group_by="ticker")
AMD['AMD']["Close"].plot(title="AMD Close Price History")
```

```python
AMD['MA30']  = AMD['AMD', 'Close'].rolling(window=30).mean()
AMD['MA100'] = AMD['AMD', 'Close'].rolling(window=100).mean()
AMD['MA200'] = AMD['AMD', 'Close'].rolling(window=200).mean()

AMD = yf.download("AMD", start=sd, end=ed, group_by="ticker")
AMD['AMD']["Close"].plot(title="AMD Price to Rolling Mean", label='Price')

windows=[30,100,200]

close = AMD['AMD']['Close']

for w in windows:
    AMD[f'MA{w}'] = close.rolling(window=w).mean()
    AMD[f'MA{w}'].plot(label=f'MA{w}')
plt.legend()
plt.show()
```

```python
AMD = yf.download("AMD", sd, ed, group_by="ticker")

AMD.columns = AMD.columns.droplevel(0)
AMD1 = AMD[AMD.index > "2023-10-01"]
AMD1.index = pd.to_datetime(AMD1.index)
mpf.plot(AMD1, title="AMD Price and Volume", type='candle', mav=(30,100,200), style='charles', volume=True)
```
